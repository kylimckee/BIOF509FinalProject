```python
#Import numpy and pandas.
import numpy as np
import pandas as pd

#Read the data and head it to make sure you have properly read it.
cc_data = pd.read_csv("cc_data.csv")
cc_data.head(5)
```

```python
#Understand the shape of the dataset you are working with.
print('The shape of the data is:', cc_data.shape)
```

```python
#Calculate statistics for each column of the dataset.
cc_data.describe()
```

```python
X = cc_data["ca_cervix"]
Y = cc_data.drop(columns=["ca_cervix"])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(Y, X, train_size=0.75, random_state=42)
```

```python
print('Training Features Shape:', X_train.shape)
print('Training Labels Shape:', Y_train.shape)
print('Testing Features Shape:', X_test.shape)
print('Testing Labels Shape:', Y_test.shape)
```

```python
display(X_train)
display(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier

trees = []
for d in [2,3,4,5]:
    trees.append(DecisionTreeClassifier(random_state=42, max_depth=d).fit(X_train, Y_train))
```

```python
def score(tree, x, y):
    return (tree.score(x, y))
for t in trees:
    print("For max depth of {}".format(t.get_params()["max_depth"]))
    print("\tAccuracy on training set: {}".format(score(t, X_train, Y_train)))
    print("\tAccuracy on test set: {}".format(score(t, X_test, Y_test)))
```

```python
from sklearn.ensemble import RandomForestRegressor

RF = RandomForestRegressor(n_estimators=4, random_state=42)
RF.fit(X_train, Y_train)
```
